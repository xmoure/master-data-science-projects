## General Data Cleaning steps

#### Removing duplicated

```python
# duplicates = df.duplicated(subset=['Track','Artist'], keep='first')
duplicates = df.duplicated(subset=['Uri'], keep='first')
duplicates.value_counts()
```

```python
df[duplicates]
```

```python
# df=df.drop_duplicates(subset=['Track','Artist'], keep='first')
df=df.drop_duplicates(subset=['Uri'], keep='first')
df.shape
```

```python
df["Licensed"]=df[df["Licensed"].isnull()==False]["Licensed"].astype(int)
df["official_video"]=df[df["official_video"].isnull()==False]["official_video"].astype(int)
```

#### Removing rows with missing values in the target variable 

```python
null_count = df['Stream'].isnull().sum()
print("Stream:", null_count)
```

```python
indexes=df[df['Stream'].isnull()].index
df=df.drop(index=indexes)
df.shape
```

```python
df.to_csv("cleaned_df1.csv", encoding='utf-8', index=False)
```
