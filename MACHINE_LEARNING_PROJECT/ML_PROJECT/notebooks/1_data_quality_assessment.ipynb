### 1.1 Data quality assessment and data exploration

Here we are going to perform some data exploration to familiarize ourselves with the dataset and to uncover meaningful insights from it.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np
import pandas as pd
```

```python
#Visualization
import plotly.express as px
import plotly.offline as pyo
import plotly.figure_factory as ff
import plotly.subplots as sp
import plotly.graph_objs as go
```

```python
df=pd.read_csv('Spotify_Youtube.csv',index_col=0)
```

#### Number of null values per each column

```python
df.isnull().sum()
```

```python
# calculate the fraction of null values for each column
null_fractions = df.isnull().mean()

# print the percentage of null values for each column
print(null_fractions * 100)
```

Duplicates

We consider that two elements are duplicates if they have the same Uri

```python
# check for duplicates
duplicates = df.duplicated(subset=['Uri'], keep='first')
duplicates.value_counts()
```

#### Descriptive statistics

```python
# Take only the numeric columns
numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()
descriptive_stats = df[numeric_columns].describe()
descriptive_stats
```

Get the top 10 artists with the most views on YouTube and streams on Spotify

```python
# Create two separate DataFrames for views and streams
df_views = df.groupby('Artist')['Views'].sum().sort_values(ascending=False)[:10]
df_streams = df.groupby('Artist')['Stream'].sum().sort_values(ascending=False)[:10]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6))

# top 10 yt
ax1.set_title('Top 10 Artists on YouTube')
df_views.plot(kind='bar', ax=ax1)

# top 10 spotofy
ax2.set_title('Top 10 Artists on Spotify')
df_streams.plot(kind='bar', ax=ax2)


ax1.set_xlabel('Artist')
ax1.set_ylabel('Views')
ax2.set_xlabel('Artist')
ax2.set_ylabel('Streams')
fig.tight_layout()
plt.show()
```

Top songs on Youtube(in views)

Top songs on Spotify

```python
top_songs = df.sort_values('Stream', ascending=False).head(10)
top_songs[['Track', 'Energy', 'Danceability', 'Acousticness']]
```

#### Correlations

Here we want to see the correlation between the variables.

We can see that the most positive correlated variables to our target (Stream) are Views and Likes. This was expected since usually songs that have a big amount of streams tend to have a lot of views in their vides and and also a lot of likes.

Then we can see that Loudness is highly correlated with Energy, which was also expected. On the other hand, we can see that the more acoustic a song is the less energy it has which makes sense. 

```python
# Calculate correlations
correlations = df[numeric_columns].corr()

# Create a correlation heatmap with custom appearance
fig = ff.create_annotated_heatmap(
    z=correlations.values,
    x=numeric_columns,
    y=numeric_columns,
    annotation_text=correlations.round(2).values,
    colorscale='Blackbody',
    showscale=True,
    hoverinfo='z'
)

# Customize the appearance
fig.update_xaxes(title_text='Features', side='bottom')
fig.update_yaxes(title_text='Features', side='left')

# Update layout 
fig.update_layout(
    title='Correlation Heatmap',
    title_x=0.5,
    width=1000,
    height=800,
    xaxis_showgrid=False,
    yaxis_showgrid=False,
    xaxis_zeroline=False,
    yaxis_zeroline=False,
    margin=dict(l=150, r=100, t=100, b=100)
)


# Display the heatmap using  the 
pyo.iplot(fig)
```

See which variables are more correlated to our target (Stream)

```python
# See the most correlated variables to Stream
corr = df.corr()
print("Most correlated to Stream")
corr['Stream'].sort_values(ascending=False)
```

Plots

From the following plots we can see that some variables like Stream, Views,Speechiness, etc. are left skewed. So, we would need to take a closer look at them when evaluating possible transformations.

```python
# variables we want to create the plot
vars_to_plot = ['Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo', 'Duration_ms', 'Views', 'Likes', 'Comments', 'Stream']

# 5x3 grid of subplots
fig, axs = plt.subplots(nrows=5, ncols=3, figsize=(15, 20))

for i, var in enumerate(vars_to_plot):
    row = i // 3
    col = i % 3
    sns.distplot(df[var], ax=axs[row, col])
    axs[row, col].set_title(var)

# display the plots
plt.tight_layout()
plt.show()
```

Scatter between Views and Streams
