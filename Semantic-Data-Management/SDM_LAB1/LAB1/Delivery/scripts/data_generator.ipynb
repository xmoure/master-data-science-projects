```python
def load_csv_into_data_frame(filename, header_filename=None):
    assert isinstance(filename, str)
    assert '.csv' in filename
    assert os.path.exists(filename)
    assert header_filename is None or '.csv' in header_filename

    # Read the header file (if provided) and extract the column names
    if header_filename is not None:
      with open(header_filename, 'r') as f:
        header = f.readline().strip().split(';')
        header = [h.split(':')[0] for h in header]
    else:
      header = None

    # Read the csv file into the pandas dataframe
    df = pd.read_csv(filename, sep=';', names=header, error_bad_lines=False)

    # Drop any rows that contain only missing values
    df.dropna(axis=0, how='all', inplace=True)

    return df
```

```python
# set the random seed
np.random.seed(12345)

n = np.random.randint(150, 350) 
data_sample = data.sample(n)
separator = ';'
# add database kewywords to the index keywords 
for index,row in data_sample.iterrows():
  keywords = row['Index Keywords']
  keywords_to_add = random.sample(database_keywords, k=3 )
  string_keywords = separator.join(keywords_to_add)
  all_keywords = keywords + ' ;' + string_keywords
  data.loc[data['PaperId'] == row['PaperId'], 'Index Keywords'] = all_keywords
```

```python
all_keywords = list()
for index,row in data.iterrows():
  keywords = [s.strip() for s in row['Index Keywords'].split(";")]
  all_keywords = np.unique(np.concatenate((all_keywords, keywords )))


data_keywords = pd.DataFrame(all_keywords, columns=['keyword'])

data_keywords['keywordId'] = [str(uuid.uuid4()) for _ in range(len(data_keywords))]

cols = ['keywordId', 'keyword']
data_keywords = data_keywords.reindex(columns= cols)

if data_keywords['keyword'].duplicated().any():
    print('The keyword column has repeated values')
else:
  print('No repetitions')
```

```python
has_key_dic = {}
for index,row in data.iterrows():
  keywords = row['Index Keywords'].split(";")
  for ind, keyword in enumerate(keywords):
    keyword = keyword.strip()
    index = data_keywords.loc[data_keywords['keyword'] == keyword].index[0]
    element = data_keywords.iloc[index]
    if(element['keywordId'] in has_key_dic):
      has_key_dic[element['keywordId']].append(row['PaperId'])
    else:
      has_key_dic[element['keywordId']]= [row['PaperId']]
```

```python
has_key_dataf = pd.DataFrame(columns=['PaperId', 'KeywordId' ])
for key,value in has_key_dic.items():
  if(len(value)==1):
    new_row = pd.DataFrame({'PaperId': [value[0]], 'KeywordId': [key]})
    has_key_dataf = pd.concat([has_key_dataf, new_row ], ignore_index=True)
  else:
    for ind, paper in enumerate(value):
      new_row =pd.DataFrame({'PaperId': [value[ind]], 'KeywordId': [key]})
      has_key_dataf = pd.concat([has_key_dataf, new_row ], ignore_index=True)

has_key_dataf= has_key_dataf.drop_duplicates()
```

```python
# author writes paper relationship and create  affiliations
university_keywords = ['college, academy', 'university', 'institute']
author_writes_dataf = pd.DataFrame(columns=['AuthorId', 'PaperId', 'IsCorresponding' ])
affiliation_dataf =  pd.DataFrame(columns=['AuthorId', 'OrganizationName', 'Type' ])
#lead_author_df = pd.DataFrame(columns=['AuthorId', 'PaperId' ])
for index,row in data.iterrows():
  authors = row['Author(s) ID'].split(";")
  authors = list(filter(lambda x: x != "", authors))
  paperId = row['PaperId']
  for ind, author in enumerate(authors):
      # Assign the first author as main author
      corresponding = False
      if(ind == 0):
        corresponding = True
      new_row = pd.DataFrame({'AuthorId': [author], 'PaperId': [paperId], 'IsCorresponding': bool(corresponding)})
      author_writes_dataf = pd.concat([author_writes_dataf, new_row ], ignore_index=True)

      #create affiliations
      affiliations = row['Affiliations'].split(";")

      i_index = ind if (len(affiliations) == len(authors)) else 0
      type_affiliation = 'University' if any(element.lower() in affiliations[i_index].lower() for element in university_keywords) else 'Company'
      new_row_aff = pd.DataFrame({'AuthorId': [author], 'OrganizationName': [affiliations[i_index]], 'Type': type_affiliation})
      affiliation_dataf = pd.concat([affiliation_dataf, new_row_aff ], ignore_index=True)

#remove duplicates
affiliation_dataf = affiliation_dataf.drop_duplicates(subset=['AuthorId'])
affiliation_dataf = affiliation_dataf[affiliation_dataf['AuthorId'].isin(authors_data_frame['AuthorId'])]


author_writes_dataf = author_writes_dataf.drop_duplicates(subset=['AuthorId', 'PaperId'])
```

```python
np.random.seed(123)
citations_dataf = pd.DataFrame(columns=['PaperId', 'CitedPaper'])
for index, paper in data.iterrows():
  #get other papers which year is minor or equal to the actual paper
  possible_papers = data[data['Year'] <= paper['Year']]['PaperId']
  possible_papers = possible_papers[possible_papers != paper['PaperId']]
  cited_papers = []
  if len(possible_papers) < 15:
    cited_papers = possible_papers['PaperId']
  else:
    num_citations = min(random.randint(15, 30), len(possible_papers))
    cited_papers = random.sample(list(possible_papers), num_citations)
  for ind, cited_paper in enumerate(cited_papers):
      new_row = pd.DataFrame({'PaperId': [paper['PaperId']], 'CitedPaper': [cited_paper]})
      citations_dataf  = pd.concat([citations_dataf , new_row ], ignore_index=True)


citations_dataf = citations_dataf.drop_duplicates(subset= ['PaperId', 'CitedPaper'])
```

```python
#Create editions

editions = pd.DataFrame(columns=['Name', 'Year', 'ConferenceId'])
for ind, conference in conferences.iterrows():
  first_three_chars = conference['Name'][:3]
  random_num = random.randint(1, 8)
  i= 1
  while i<=random_num:
    random_string = ''.join(random.choices(string.ascii_uppercase + string.digits, k=6))
    new_edition_name = first_three_chars + random_string
    #conference_year = conference['Year']
    #random_year = random.randint(conference_year, 2021)
    new_row = pd.DataFrame({'Name': [new_edition_name], 'Year': [conference['Year']], 'ConferenceId': [conference['ConferenceId']]})
    editions = pd.concat([editions, new_row ], ignore_index=True)
    i+= 1
#create edition id
editions['EditionId'] = [str(uuid.uuid4()) for _ in range(len(editions))]

#create relationship between edition and conference
edition_is_part_of_conference = pd.DataFrame(columns=['EditionId', 'ConferenceId'])
for ind, edition in editions.iterrows():
  new_edge = pd.DataFrame({'EditionId': [edition['EditionId']], 'ConferenceId': [edition['ConferenceId']]})
  edition_is_part_of_conference = pd.concat([edition_is_part_of_conference, new_edge ], ignore_index=True)

editions
```
