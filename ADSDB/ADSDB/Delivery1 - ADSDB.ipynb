```python
import pandas as pd
import numpy as np
import pycountry
```

```python
chocolate = pd.read_csv("drive/MyDrive/ADSDB/landing/persistent/chocolate.csv")
freedom = pd.read_csv("drive/MyDrive/ADSDB/landing/persistent/freedom.csv")
```

```python
chocolate.company_location = chocolate.company_location.str.replace(".", "")
```

```python
def search_country(x):
  try:
    return pycountry.countries.search_fuzzy(x)[0].alpha_2
  except:
    return "ERROR"
  
chocolate["alpha2_company_location"] = chocolate.company_location.apply(lambda x : search_country(x))
```

```python
chocolate["alpha2_country_of_bean_origin"] = chocolate.country_of_bean_origin.apply(lambda x : search_country(x))
freedom["alpha2_country"] = freedom.country.apply(lambda x : search_country(x))
```
